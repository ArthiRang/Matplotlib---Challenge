```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_metadata = pd.merge(study_results,mouse_metadata, how = "left", on=["Mouse ID","Mouse ID"])

# Display the data table for preview

study_metadata.head()

```

```python
# Checking the number of mice.
number_of_mice=len(study_metadata["Mouse ID"].unique())
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicateMouseID = study_metadata.loc[study_metadata.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicateMouseID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dupmouseID = study_metadata.loc[study_metadata["Mouse ID"] == "g989"]
dupmouseID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
StudyDataClean = study_metadata[study_metadata['Mouse ID'] !="g989" ]
StudyDataClean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
newMouse = len(StudyDataClean["Mouse ID"].unique())
newMouse
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
StudyDataClean_Mean =StudyDataClean.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]

StudyDataClean_Median =StudyDataClean.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]

StudyDataClean_Variance =StudyDataClean.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]

StudyDataClean_StandardDeviation =StudyDataClean.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]

StudyDataClean_SEM =StudyDataClean.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]


summaryStatisticsdf = pd.DataFrame({'Mean Tumor Volume': StudyDataClean_Mean,
                                    'Median Tumor Volume':StudyDataClean_Median,
                                    'Tumor Volume Variance':StudyDataClean_Variance,
                                    'Tumor Volume Std. Dev.':StudyDataClean_StandardDeviation,
                                    'Tumor Volume Std. Err.':StudyDataClean_SEM
                                   
                                   })
summaryStatisticsdf
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

aggsummaryStatisticsdf =  StudyDataClean.groupby("Drug Regimen")["Tumor Volume (mm3)"].aggregate(["mean",
                                                                                                      "median",
                                                                                                      "var",
                                                                                                      "std",
                                                                                                      "sem"])

aggsummaryStatisticsdf
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

new_Data =StudyDataClean.groupby(["Mouse ID"])["Timepoint"].max()
mouseData = new_Data.reset_index()
mouse_Data_New = pd.merge( mouseData,StudyDataClean, how = "left", on=["Mouse ID","Timepoint"])
mouse_Data_New 
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumor vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
    
treatment_list = ['Capomulin','Ramicane','Infubinol','Ceftamin']

volData = []
for treatment in treatment_list:
    new_Study_Data_Final = mouse_Data_New.loc[mouse_Data_New["Drug Regimen"]== treatment,"Tumor Volume (mm3)"]
    
    #print(new_Study_Data_Final)
    
    volData.append(new_Study_Data_Final)
    
    quartiles = new_Study_Data_Final.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq   
    lowerBound = lowerq - (1.5*iqr)
    upperBound = upperq + (1.5*iqr)
    outliers= new_Study_Data_Final.loc[(new_Study_Data_Final < lowerBound)|(new_Study_Data_Final > upperBound)]
    
    
    print(f"{treatment}'s Potential Outliers:{outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
propcol = dict(markerfacecolor ='red',markersize = 10)
ax1.boxplot(volData, labels = treatment_list,flierprops = propcol)
plt.show()

```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_Data = StudyDataClean[(StudyDataClean["Drug Regimen"] == "Capomulin") & (StudyDataClean["Mouse ID"] == "l509")]

#capomulin_Data

fig1,ax1 = plt.subplots(figsize= (5,3))
ax1.set_title('Capomulin treatment of mouse l509')
ax1.set_xlabel('Timepoint(days)')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.plot(capomulin_Data["Timepoint"],capomulin_Data["Tumor Volume (mm3)"])
plt.show()

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
avg_Tumor_Vol = StudyDataClean[(StudyDataClean["Drug Regimen"] == "Capomulin")].groupby(["Mouse ID"])[["Weight (g)","Tumor Volume (mm3)"]].mean()                        
                                   
#avg_Tumor_Vol                                   
                                   
fig1,ax1 =plt.subplots()

plt.scatter(avg_Tumor_Vol["Weight (g)"],avg_Tumor_Vol["Tumor Volume (mm3)"])
ax1.set_xlabel('Weight (g)')
ax1.set_ylabel('Average Tumor Volume (mm3)')
plt.show()


```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

x_values = avg_Tumor_Vol['Weight (g)']
y_values = avg_Tumor_Vol['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y =" + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8), color ="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
